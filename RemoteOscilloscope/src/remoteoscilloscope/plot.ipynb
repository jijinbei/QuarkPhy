```python
import matplotlib.pyplot as plt
import polars as pl
```

```python
PATH = "../../20240703.csv"
RECORD_LENGTH = 250000
```

```python
def plot(ax, x, y, label, color="b"):
    ax.plot(x, y, label=label, color=color)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Voltage [V]")
    ax.legend()
    ax.grid()
```

```python
all_df = pl.read_csv(PATH)

dfs = []
for i in range(0, len(all_df), RECORD_LENGTH):
    dfs.append(all_df.slice(i, RECORD_LENGTH))
```

```python
num = 1
fig, ax = plt.subplots(nrows=2)

plot(ax[0], dfs[num]['TIME'], dfs[num]['CH1'], 'CH1', color='b')
plot(ax[1], dfs[num]['TIME'], dfs[num]['CH2'], 'CH2', color='r')

plt.show()
```
